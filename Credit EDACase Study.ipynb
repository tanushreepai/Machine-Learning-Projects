# Business Problem Statement

The loan providing companies find it hard to give loans to the people due to their insufficient or non-existent credit history. Because of that, some consumers use it as their advantage by becoming a defaulter. Suppose we work for a consumer finance company which specialises in lending various types of loans to urban customers. We have to use EDA to analyse the patterns present in the data. This will ensure that the applicants are capable of repaying the loan are not rejected.


<span class="girk"><span class="mark">We need to identify patterns which indicate if a client has difficulty paying their instalments which may be used for taking actions such as denying the loan, reducing the amount of loan, lending (to risky applicants) at a higher interest rate, etc. This will ensure that the consumers capable of repaying the loan are not rejected.</span></span> 

Our Aim is to identification of such applicants using this EDA.

In other words, the company wants to understand the driving factors (or driver variables) behind loan default, i.e. the variables which are strong indicators of default.  The company can utilise this knowledge for its portfolio and risk assessment.


# Common Reasons for a Loan Rejection

## <font color='blue'>Denial Due to Credit Score</font>
If there were too many applications/rejections previously it impacts Credit Score.
Previous late payments or any kind of bankruptcy previously affects the loan sanction decision.
A thin credit history means that if a person does not hold any financial products – i.e.Credit Cards, loans.  In this case, lenders cannot assess the  financial behaviour, and therefore, are hesitant to sanction the loan

## <font color='blue'>Insufficient Income</font>
Those who do not make enough money, and cannot document other sources of income, are frequently denied. Those who have lost a job, or even recently started a new job, may be denied as well.
 Self-employed applicants are usually asked to submit evidence documenting a steady flow of business.
 
## <font color='blue'>Job instability including Constant job hopping</font>
Switching jobs within 6 months regularly attracts rejection possibility of loan

## <font color='blue'>An Abundance of Debt i.e. Debt income ratio too high</font>
If there is high Credit card utilization or there are too many loans then possibility of loan rejection arises

## <font color='blue'>Poor Documentation i.e. Incomplete or incorrect loan application</font>
Failure to submit documents or signature mismatch or wrong information in submitted documents leads to rejection.

### Outlier Handling

For outlier handling we have deployed the IQR Method

```python
# function to remove the outliers using IQR method.

def outliers_IQR(data, col):
    lower_quartile = data[col].quantile(0.25)
    upper_quartile = data[col].quantile(0.75)
    IQR = upper_quartile - lower_quartile
    outlier_thresh = 1.5 * IQR
    return data[data[col].between((lower_quartile - outlier_thresh), (upper_quartile + outlier_thresh))]
```

## Univariate and Bivariate Analysis of Categorical Variables

```python
# Identifying the categorical variables

ca.select_dtypes('object').apply(pd.Series.nunique, axis = 0)
```

```python
# Function to calculate the average default percentage across the categories
# It can also compute the 'difference between the highest and the lowest default rate' across the 
# categories, which can be used as a parameter to indicate the effect of the varaible on default rate

def default_probability(check):
    default_across = ca.groupby(check).TARGET.mean().sort_values(ascending=False)
    return (round(((default_across[0] - default_across[-1])*100),2))
```
