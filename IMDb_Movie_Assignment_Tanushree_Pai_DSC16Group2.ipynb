-  ### Subtask 3.6: Find popular genres

You might have noticed the `genres` column in the dataframe with all the genres of the movies seperated by a pipe (`|`). Out of all the movie genres, the first two are most significant for any film.

1. Extract the first two genres from the `genres` column and store them in two new columns: `genre_1` and `genre_2`. Some of the movies might have only one genre. In such cases, extract the single genre into both the columns, i.e. for such movies the `genre_2` will be the same as `genre_1`.
2. Group the dataframe using `genre_1` as the primary column and `genre_2` as the secondary column.
3. Find out the 5 most popular combo of genres by finding the mean of the gross values using the `gross` column and store them in a new dataframe named `PopGenre`.

```python
# Extracting the first two genres from the genres column and storing them in new columns genres_1 and genres_2

movies['genres_1']= movies['genres'].str.split('|').str[0]
movies['genres_2']= movies['genres'].str.split('|').str[1] 

# If there is only one genres in original dataframe then assigning the same to genres_1 and genres_2 by replacing
# genres_2 null values as obtained above with genres_1 value
movies.loc[pd.isnull(movies['genres_2']), ['genres_2']] = movies['genres_1']
movies
```

```python
# Grouping the dataframe by genres_1 and genres_2 and saving in new dataFrame named 'movies_by_segment'

movies_by_segment = movies.groupby(['genres_1','genres_2'])
```

```python
# Obtaining the 5 most popular combo of genres here using the 'movies_by_segment' DataFrame of the previous step
# Sorting in descending order by mean of the gross values

PopGenre= pd.DataFrame(movies_by_segment['gross'].mean()).sort_values('gross',ascending=False).head(5)
PopGenre
```

Create three new dataframes namely, `Meryl_Streep`, `Leo_Caprio`, and `Brad_Pitt` which contain the movies in which the actors: 'Meryl Streep', 'Leonardo DiCaprio', and 'Brad Pitt' are the lead actors. Use only the `actor_1_name` column for extraction. Also, make sure that you use the names 'Meryl Streep', 'Leonardo DiCaprio', and 'Brad Pitt' for the said extraction.
    2. Append the rows of all these dataframes and store them in a new dataframe named `Combined`.
    3. Group the combined dataframe using the `actor_1_name` column.
    4. Find the mean of the `num_critic_for_reviews` and `num_users_for_review` and identify the actors which have the highest mean.

###### Creating 3 new DataFrames Meryl_Streep, Leo_Caprio, and Brad_Pitt

```python
# Creating new dataframe named Meryl_Streep by extracting rows of movies DataFrame were lead actor is 'Meryl Streep'

Meryl_Streep = movies.loc[movies.actor_1_name=='Meryl Streep']
Meryl_Streep
```

```python
# Creating new dataframe named Leo_Caprio by extracting rows of movies DataFrame were lead actor is 'Leonardo DiCaprio'

Leo_Caprio = movies.loc[movies.actor_1_name=='Leonardo DiCaprio']
Leo_Caprio
```

```python
# Creating new dataframe named Brad_Pitt by extracting rows of movies DataFrame were lead actor is 'Brad Pitt'

Brad_Pitt = movies.loc[movies.actor_1_name=='Brad Pitt']
Brad_Pitt
```

```python
# Creating "Combined" Dataframe with all the values of 'Meryl_Streep','Leo_Caprio' & 'Brad_Pitt' DataFrames appended

Combined = pd.concat([Meryl_Streep, Leo_Caprio, Brad_Pitt],axis=0)
Combined
```

```python
# Grouping the combined dataframe with Lead actor column i.e. actor_1_name

Combined.groupby('actor_1_name')
```

```python
# Checking if the columns 'num_critic_for_reviews' & 'num_user_for_reviews' have null values as mean will be calculated next

Combined.isnull().sum()
```

```python
# Finding the mean of critic reviews and audience reviews using pivot table and then sorting in descending order

Combined_mean=pd.DataFrame(Combined.pivot_table(values=['num_critic_for_reviews','num_user_for_reviews'],
                              index='actor_1_name',
                              aggfunc='mean')).sort_values(['num_critic_for_reviews','num_user_for_reviews'],ascending=False)
Combined_mean
```

**Checkpoint 6:** `Leonardo` has aced both the lists!

```python
# Adding a new column 'decade' in the movies DataFrame consisting of the decade in which each movie was released

movies['decade'] = movies['title_year'].astype(str).str[0:3]+'0s'    
movies
```

```python
# Creating a new dataframe 'df_by_decade' where movies are grouped by 'decade' and sum of no of users voted is calculated 

df_by_decade=pd.DataFrame(movies.groupby('decade',as_index=False).num_voted_users.sum())
df_by_decade
```

```python
# Plotting number of voted users vs decade using barplot

plt.figure(figsize=(12,9))
graph=sns.barplot(x='decade',y='num_voted_users',data=df_by_decade)
plt.title("Voted Users vs Decade")
plt.xlabel("Decades")
plt.ylabel("No of Voted Users (in log scale)")
# Converting the no of voted users scale to logarithmic for better visualization
plt.yscale("log")

# Values on top of each bar shows the original values of the number of voted users although the bar plot visualization scale is logarithmic
for index, row in df_by_decade.iterrows():
    graph.text(row.name,row.num_voted_users, (row.num_voted_users), color='black', ha="center",va="bottom")
plt.show()
```

#### Analysis

-  It is evident from the graph the number of users who voted for a movie released in 1920s is far less than the number of users who voted for movies released in the 2010s decade. 
-  There has been a continous growth in the number of voted users across the decades except decade 1930s being an  exceptionally high number of voted users as compared to its adjacent decades. Probably movies provided an escape from the hardships of the Great Depression of 1929.So more people watched and voted for movies.
-  An important point to be considered here is the data for decade 2010 is only till 2016, so when the 3 more years of the decade will be considered the no of voted users may be more than that of 2000s decade following the growth pattern of the graph.

The bar plot is plotted with the number of voted users value converted to logarithmic scale for better visulaization as if the normal no of users is plotted the initial 4 years are not visible due to large differnce in lowest and highest values. Each bar for every decade has been labelled with the actual no of voted users on the top of bar.
